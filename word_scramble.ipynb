Word Unscrambler
---

In this program, we are going to create a word unscrambler. 

First we need to find a text file that has all the possible words in the English dictionary. I saved this file as "words.txt".

In order to access the file, I had to access the file using the open function:

```python
wordfile = open("words.txt", "r")
```

---

I then wanted to add the words in this file to a list. I first was required to create an empty list, which I called word_list:

```python
word_list = [] 
words = wordfile.readlines()  #reads and returns a list of lines from file
```
---

In the next section, I wanted to go through each word on each line and perform multiple actions: 
1. use the lower() function to ensure all my words were lower case
2. use the strip() function to eliminate the extra spaces before and after my word, as well as deleting the \n at the end
3. after the words were "cleaned up" to my liking, I added them to my word_list


```python
for word in words:
    word= word.lower() #ensures words are all lower case
    word = word.strip() #deletes the extra spaces before/after a word and the \n 
    word_list.append(word) #adds the word to the list

word_set = set(word_list) #set will delete any duplicates
word_set= sorted(word_set) #sorts alphabetically
word_list = word_set
#print(word_list)
```

After I had the list, I wanted to make sure that identical words were on my list more than once, so I used the set() function to eliminate the extras. I then sorted the words alphabetically and set it equal to the word_list to update my values.

The code above places some restrictions on the user's word. It must be longer than 3 characters and if it is not, the user will have to continuously edit their word until it is greater than 3.

The if statement will splice the word down to the first 6 characters if it is longer than 6 characters. Since we are using the index values, we will start at index 0 and end at index 5.

---

The next line of code will break the entered word into its characters:

```python
sw_sorted= sorted(selected_word) #sorts into individual characters
```

---

The next section creates a new list of our possible words. These are words that contain at least one character that is the same as the entered word:

```python
possible_words = []
for word in word_list:
    for ch in word:
        if ch in sw_sorted:
            possible_words.append(word)
```

It goes through each word on the word list, checking each character in the word and comparing it to the characters from our selected word. If it satisfies these conditions it is added to our new list.

I will post the entire code below so you can see it all in one place

````python
#task 1: decomposing money
entry = int(input("Enter a number 1-9999: "))
while entry <1 or entry>9999:
    print("Invalid entry. Please enter a number within the range.")
    entry = int(input("Enter a number 1-9999"))
    
num_grands = int(entry/1000)
remain = entry % 1000
print(num_grands, " grands")

num_hundreds = int(remain/100)
remain %= 100
print(num_hundreds, " Benjamins")

num_tens = int(remain/10)
remain %= 10
print(num_tens, " sawbucks")

print(remain, " bucks")

    
#task 2: unscramble words
wordfile = open("words.txt","r")
word_list = [] #creates an empty list for our words

words = wordfile.readlines()  #reads and returns a list of lines from file

for word in words:
    word= word.lower() #ensures words are all lower case
    word = word.strip() #deletes the extra spaces before/after a word and the \n 
    word_list.append(word) #adds the word to the list

word_set = set(word_list) #set will delete any duplicates
word_set= sorted(word_set) #sorts alphabetically
word_list = word_set
#print(word_list)

#allows user to select a word
selected_word = input("Enter a word: ")
while len(selected_word) < 3: #invalid length 
    selected_word = input("Invalid word length. Please enter a new word: ")
    if len(selected_word)>3:
        break

if len(selected_word)>6: #makes alterations if the length is too long
    selected_word = selected_word[0:5]

sw_sorted= sorted(selected_word) #sorts into individual characters
#print(sw_sorted)

possible_words = []
for word in word_list:
    for ch in word:
        if ch in sw_sorted:
            possible_words.append(word)

possible_words = set(possible_words)
pw_sorted = sorted(possible_words, key=len)
pw_6 = []
pw_5 = []
pw_4 = []
pw_3 = []

for word in pw_sorted:
    if len(word) == 6:
        pw_6.append(word)
    elif len(word) == 5:
        pw_5.append(word)
    elif len(word) == 4:
        pw_4.append(word)
    elif len(word) == 3:
        pw_3.append(word)


if len(selected_word) == 6: 
    print("\n\nThese are all the possible 6 letter words: ")
    for word in pw_6:
        print(word, end = ", ")
if len(selected_word) == 5 or len(selected_word) == 6:
    print("\n\nThese are all the possible 5 letter words: ")
    for word in pw_5:
        print(word, end = ", ")
if len(selected_word) >= 4 and len(selected_word) <=6: 
    print("\n\nThese are all the possible 4 letter words: ")
    for word in pw_4:
        print(word, end= ", ")
if len(selected_word) >= 3 and len(selected_word) <=6:
    print("\n\nThese are all the possible 3 letter words: ")
    for word in pw_3:
        print(word, end = ", ")
````
